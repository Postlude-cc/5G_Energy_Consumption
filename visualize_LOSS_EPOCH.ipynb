```python
import matplotlib.pyplot as plt

with open(f"./Energy_Consumption_Model_progress.txt", "r") as f:
    lines = f.readlines()

losses = []
epoch_durations = []

for line in lines:
    parts = line.split(',')
    losses.append(float(parts[0].split(':')[-1].strip()))
    epoch_durations.append(float(parts[1].split(':')[-1].strip().split()[0]))

epochs = list(range(1, len(losses)+1))

plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
plt.plot(epochs, losses, label='Loss', color='red')
plt.title('Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(epochs, epoch_durations, label='Epoch Duration', color='blue')
plt.title('Epoch Duration over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Duration (seconds)')
plt.legend()

plt.tight_layout()
plt.show()
```
